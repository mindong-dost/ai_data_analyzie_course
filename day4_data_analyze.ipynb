```python
# =================================================================
# ì´ ì…€ì€ ì ˆëŒ€ ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”.
# í•¨ìˆ˜ì™€ ì½”ë“œ ì„¤ì •ì´ í¬í•¨ë˜ì–´ ìˆìŠµë‹ˆë‹¤.
# =================================================================

MAX_VIDEOS = 10             # ë¶„ì„í•  ì˜ìƒ ê°œìˆ˜ (10ê°œ)
MAX_RAW_VIDEOS = 500       # ì¸ê¸°ìˆœ í˜¹ì€ ìµœì‹ ìˆœ ì •ë ¬ì„ ìœ„í•´ ê°€ì ¸ì˜¬ ë¹„ë””ì˜¤ ê°œìˆ˜
MAX_COMMENTS_PER_VIDEO = 20 # ì˜ìƒë‹¹ ëŒ“ê¸€ ê°œìˆ˜ (20ê°œ)

# ê²°ê³¼ ì €ì¥í•  í´ë” ì„¤ì •
SAVE_PATH = "/content/drive/MyDrive/youtube_analysis/"
os.makedirs(SAVE_PATH, exist_ok=True)  # í´ë”ê°€ ì—†ìœ¼ë©´ ìë™ ìƒì„±

print("ì„¤ì • ì™„ë£Œ!")
print(f"ì €ì¥ ìœ„ì¹˜: {SAVE_PATH}")


# =================================================================
# API ì—°ê²° ì„¤ì •
# =================================================================

# YouTube Data API ì—°ê²°
try:
    youtube = build('youtube', 'v3', developerKey=YOUTUBE_API_KEY)
    print("YouTube API ì—°ê²° ì„±ê³µ!")
except Exception as e:
    print(f"YouTube API ì—°ê²° ì‹¤íŒ¨: {e}")
    print("API í‚¤ë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”")

# Gemini AI ì—°ê²°
try:
    os.environ["GEMINI_API_KEY"] = GEMINI_API_KEY
    genai_client = genai.Client()
    print("Gemini AI ì—°ê²° ì„±ê³µ!")
except Exception as e:
    print(f"Gemini AI ì—°ê²° ì‹¤íŒ¨: {e}")
    print("API í‚¤ë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”")



def get_channel_id_by_handle(handle):
    """ì±„ë„ í•¸ë“¤ë¡œ ì‹¤ì œ ì±„ë„ IDë¥¼ ì°¾ëŠ” í•¨ìˆ˜"""
    try:
        handle = handle.replace('@', '')
        request = youtube.search().list(
            part='snippet',
            q=handle,
            type='channel',
            maxResults=1
        )
        response = request.execute()
        if response['items']:
            return response['items'][0]['snippet']['channelId']
        return None
    except Exception as e:
        print(f"ì±„ë„ ID ì°¾ê¸° ì—ëŸ¬: {e}")
        return None

def get_all_channel_videos(youtube, channel_id, limit=MAX_RAW_VIDEOS):
    """
    ì§€ì •ëœ ì±„ë„ì˜ ë™ì˜ìƒ ìƒì„¸ ì •ë³´ë¥¼ ì§€ì •ëœ ìµœëŒ€ ê°œìˆ˜(limit)ë§Œí¼ë§Œ ê°€ì ¸ì˜µë‹ˆë‹¤.

    Args:
        youtube: YouTube API ì„œë¹„ìŠ¤ ê°ì²´
        channel_id (str): ì¡°íšŒí•  ì±„ë„ì˜ ID
        limit (int): ê°€ì ¸ì˜¬ ìµœëŒ€ ë™ì˜ìƒ ê°œìˆ˜

    Returns:
        list: ì±„ë„ì˜ ë™ì˜ìƒ ì •ë³´ê°€ ë‹´ê¸´ ë¦¬ìŠ¤íŠ¸
    """
    try:
        # (ì´ì „ê³¼ ë™ì¼) ì±„ë„ì˜ 'ì—…ë¡œë“œ' ì¬ìƒëª©ë¡ ID ê°€ì ¸ì˜¤ê¸°
        channel_response = youtube.channels().list(part='contentDetails', id=channel_id).execute()
        uploads_playlist_id = channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

        video_ids = []
        next_page_token = None

        # while ë£¨í”„ ì‹œì‘
        while True:
            playlist_response = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=uploads_playlist_id,
                maxResults=50,
                pageToken=next_page_token
            ).execute()

            video_ids.extend([item['contentDetails']['videoId'] for item in playlist_response['items']])

            # --- âœ¨ í•µì‹¬ ìˆ˜ì • ë¶€ë¶„ ---
            # 1. í˜„ì¬ê¹Œì§€ ê°€ì ¸ì˜¨ ë™ì˜ìƒ ê°œìˆ˜ê°€ limitì„ ë„˜ì—ˆëŠ”ì§€ í™•ì¸
            if len(video_ids) >= limit:
                break # ë„˜ì—ˆë‹¤ë©´ ë£¨í”„ ì¤‘ë‹¨

            next_page_token = playlist_response.get('nextPageToken')

            # 2. ë‹¤ìŒ í˜ì´ì§€ê°€ ì—†ì–´ë„ ë£¨í”„ ì¤‘ë‹¨
            if not next_page_token:
                break

        # --- âœ¨ ê°€ì ¸ì˜¨ ID ê°œìˆ˜ì— ë§ì¶° ìë¥´ê¸° ---
        # limitì„ ì´ˆê³¼í•´ì„œ ê°€ì ¸ì™”ì„ ìˆ˜ ìˆìœ¼ë¯€ë¡œ ì •í™•íˆ limit ê°œìˆ˜ë§Œí¼ë§Œ ì˜ë¼ëƒ…ë‹ˆë‹¤.
        limited_video_ids = video_ids[:limit]

        all_videos = []
        for i in range(0, len(limited_video_ids), 50):
            videos_response = youtube.videos().list(
                part='snippet,statistics,contentDetails',
                id=','.join(limited_video_ids[i:i+50])
            ).execute()
            all_videos.extend(videos_response['items'])

        return all_videos

    except Exception as e:
        print(f"ë°ì´í„° ìˆ˜ì§‘ ì¤‘ ì—ëŸ¬ ë°œìƒ: {e}")
        return []

def analyze_channel_videos(videos, sort_by='popular', start_date=None, end_date=None, limit=MAX_VIDEOS):
    """
    ê°€ì ¸ì˜¨ ë™ì˜ìƒ ë¦¬ìŠ¤íŠ¸ë¥¼ ì£¼ì–´ì§„ ì¡°ê±´ì— ë”°ë¼ í•„í„°ë§í•˜ê³  ì •ë ¬í•©ë‹ˆë‹¤.

    Args:
        videos (list): ë¶„ì„í•  ì „ì²´ ë™ì˜ìƒ ë¦¬ìŠ¤íŠ¸
        sort_by (str): 'popular' (ì¸ê¸°ìˆœ), 'latest' (ìµœì‹ ìˆœ) ì¤‘ í•˜ë‚˜
        start_date (str): 'YYYY-MM-DD' í˜•ì‹ì˜ ì‹œì‘ ë‚ ì§œ
        end_date (str): 'YYYY-MM-DD' í˜•ì‹ì˜ ì¢…ë£Œ ë‚ ì§œ

    Returns:
        list: ì¡°ê±´ì— ë§ê²Œ ì²˜ë¦¬ëœ ë™ì˜ìƒ ë¦¬ìŠ¤íŠ¸
    """
    processed_videos = videos

    # 1. ë‚ ì§œ í•„í„°ë§
    if start_date and end_date:
        start_dt = datetime.fromisoformat(start_date + 'T00:00:00Z')
        end_dt = datetime.fromisoformat(end_date + 'T23:59:59Z')
        processed_videos = [
            v for v in processed_videos
            if start_dt <= datetime.fromisoformat(v['snippet']['publishedAt'].replace('Z', '+00:00')) <= end_dt
        ]

    # 2. ì •ë ¬
    if sort_by == 'popular':
        processed_videos = sorted(
            processed_videos,
            key=lambda v: int(v['statistics'].get('viewCount', 0)),
            reverse=True
        )
    elif sort_by == 'latest':
        processed_videos = sorted(
            processed_videos,
            key=lambda v: v['snippet']['publishedAt'],
            reverse=True
        )

    return processed_videos[:limit]

def convert_dataframe_from_video_data(videos):
  videos_data = []
  for video in videos:
    video_data = {
        "video_id": video.get("id"),
        "title": video.get("snippet").get("title"),
        "channel_title": video.get("snippet").get("channelTitle"),
        "upload_date": video.get("snippet").get("publishedAt"),
        "view_count": video.get("statistics").get("viewCount"),
        "like_count": video.get("statistics").get("likeCount"),
        "comment_count": video.get("statistics").get("commentCount"),
        "duration": video.get("contentDetails").get("duration"),
        "description": video.get("snippet").get("description"),
        "collected_at": datetime.now().isoformat()
    }
    videos_data.append(video_data)

  videos_df = pd.DataFrame(videos_data)
  return videos_df

def collect_video_comments(videos_df, youtube=youtube, max_comments_per_video=MAX_COMMENTS_PER_VIDEO):
    """
    ì£¼ì–´ì§„ ë™ì˜ìƒ ëª©ë¡(DataFrame)ì— ëŒ€í•´ ëŒ“ê¸€ì„ ìˆ˜ì§‘í•˜ì—¬ DataFrameìœ¼ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.

    Args:
        youtube: YouTube API ì„œë¹„ìŠ¤ ê°ì²´
        videos_df (pd.DataFrame): 'video_id', 'title' ì»¬ëŸ¼ì„ í¬í•¨í•œ ë°ì´í„°í”„ë ˆì„
        max_comments_per_video (int): ì˜ìƒë‹¹ ìˆ˜ì§‘í•  ìµœëŒ€ ëŒ“ê¸€ ìˆ˜

    Returns:
        pd.DataFrame: ìˆ˜ì§‘ëœ ëª¨ë“  ëŒ“ê¸€ ì •ë³´ê°€ ë‹´ê¸´ ë°ì´í„°í”„ë ˆì„
    """
    if videos_df.empty:
        print("ë¶„ì„í•  ì˜ìƒì´ ì—†ì–´ ëŒ“ê¸€ ìˆ˜ì§‘ì„ ê±´ë„ˆëœë‹ˆë‹¤.")
        return pd.DataFrame()

    all_comments = []
    print("\n=================================================================")
    print("ì˜ìƒ ëŒ“ê¸€ ìˆ˜ì§‘ ì‹œì‘")
    print("=================================================================")

    for idx, video in videos_df.iterrows():
        video_id = video['video_id']
        video_title = video['title']
        print(f"\n[{idx+1}/{len(videos_df)}] '{video_title[:40]}...' ëŒ“ê¸€ ìˆ˜ì§‘ ì¤‘...")

        try:
            comments_request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=max_comments_per_video,
                order='relevance'  # ì¸ê¸°ìˆœ (ë˜ëŠ” 'time'ìœ¼ë¡œ ìµœì‹ ìˆœ)
            )
            comments_response = comments_request.execute()

            video_comments = []
            for item in comments_response['items']:
                top_comment = item['snippet']['topLevelComment']['snippet']
                comment_data = {
                    'video_id': video_id,
                    'video_title': video_title,
                    'comment_id': item['snippet']['topLevelComment']['id'],
                    'text': top_comment['textDisplay'],
                    'author': top_comment['authorDisplayName'],
                    'like_count': top_comment.get('likeCount', 0),
                    'published_at': top_comment['publishedAt'],
                    'reply_count': item['snippet'].get('totalReplyCount', 0),
                    'collected_at': datetime.now().isoformat()
                }
                video_comments.append(comment_data)

            all_comments.extend(video_comments)
            print(f"  âœ… {len(video_comments)}ê°œ ëŒ“ê¸€ ìˆ˜ì§‘ ì™„ë£Œ")
            time.sleep(0.5)

        except HttpError as e:
            error_message = str(e)
            if 'commentsDisabled' in error_message:
                print("  âš ï¸ ëŒ“ê¸€ì´ ë¹„í™œì„±í™”ëœ ì˜ìƒì…ë‹ˆë‹¤.")
            elif 'quotaExceeded' in error_message:
                print("  ğŸ›‘ API ì‚¬ìš©ëŸ‰ ì´ˆê³¼! ëŒ“ê¸€ ìˆ˜ì§‘ì„ ì¤‘ë‹¨í•©ë‹ˆë‹¤.")
                break
            else:
                print(f"  âŒ API ì˜¤ë¥˜: {error_message}")
        except Exception as e:
            print(f"  âŒ ëŒ“ê¸€ ìˆ˜ì§‘ ì¤‘ ì•Œ ìˆ˜ ì—†ëŠ” ì˜¤ë¥˜: {e}")

    print(f"\nëŒ“ê¸€ ìˆ˜ì§‘ ì™„ë£Œ! ì´ {len(all_comments)}ê°œ ëŒ“ê¸€ ìˆ˜ì§‘")
    return pd.DataFrame(all_comments)

def preprocess_comments(comments_df):
    """
    ëŒ“ê¸€ ë°ì´í„°í”„ë ˆì„ì„ ë°›ì•„ ê¸°ë³¸ì ì¸ ì „ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.

    Args:
        comments_df (pd.DataFrame): ì›ë³¸ ëŒ“ê¸€ ë°ì´í„°í”„ë ˆì„

    Returns:
        pd.DataFrame: í•„ìš”í•œ ì»¬ëŸ¼ë§Œ ì„ íƒí•˜ê³  í…ìŠ¤íŠ¸ì˜ ì–‘ë ê³µë°±ì„ ì œê±°í•œ ë°ì´í„°í”„ë ˆì„
    """
    if comments_df.empty:
        print("ëŒ“ê¸€ ë°ì´í„°ê°€ ì—†ì–´ ë°ì´í„° ì •ë¦¬ë¥¼ ê±´ë„ˆëœë‹ˆë‹¤.")
        return pd.DataFrame()

    # í•„ìš”í•œ ì»¬ëŸ¼ë§Œ ì„ íƒí•˜ì—¬ ìƒˆë¡œìš´ ë°ì´í„°í”„ë ˆì„ ìƒì„±
    processed_df = comments_df[[
        'text', 'author', 'like_count', 'video_title'
    ]].copy()

    # 'text' ì»¬ëŸ¼ì˜ ì–‘ìª½ ë ê³µë°± ì œê±° (íš¨ìœ¨ì ì¸ pandas ë°©ì‹)
    processed_df['text'] = processed_df['text'].str.strip()

    print(f"ì´ {len(processed_df)}ê°œì˜ ëŒ“ê¸€ ì²˜ë¦¬ ì™„ë£Œ!")
    return processed_df

def generate_and_save_ai_report(
    genai_client,
    videos_df,
    comments_df,
    meaningful_comments,
    channel_name,
    prompt,
    save_path=SAVE_PATH,
    model_name="gemini-2.0-flash"
):
    """
    ì˜ìƒê³¼ ëŒ“ê¸€ ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ AI ë¶„ì„ ë¦¬í¬íŠ¸ë¥¼ ìƒì„±í•˜ê³  íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.

    Args:
        genai_client: Gemini AI í´ë¼ì´ì–¸íŠ¸ ê°ì²´
        model_name (str): ì‚¬ìš©í•  Gemini ëª¨ë¸ ì´ë¦„ (ì˜ˆ: 'gemini-1.5-flash')
        videos_df (pd.DataFrame): ë¶„ì„í•  ì˜ìƒ ì •ë³´ ë°ì´í„°í”„ë ˆì„
        comments_df (pd.DataFrame): ì „ì²´ ëŒ“ê¸€ ë°ì´í„°í”„ë ˆì„ (í†µê³„ìš©)
        meaningful_comments (list): AIì—ê²Œ ì „ë‹¬í•  ì£¼ìš” ëŒ“ê¸€ ë¦¬ìŠ¤íŠ¸
        channel_name (str): ë¶„ì„í•  ì±„ë„ì˜ ì´ë¦„
        view_threshold (int): ë¶„ì„ ì¡°ê±´ì— ëª…ì‹œí•  ì¡°íšŒìˆ˜ ê¸°ì¤€
        save_path (str): ë¶„ì„ ë¦¬í¬íŠ¸ íŒŒì¼ì„ ì €ì¥í•  ê²½ë¡œ

    Returns:
        str: ì €ì¥ëœ íŒŒì¼ì˜ ê²½ë¡œ. ì‹¤íŒ¨ ì‹œ None ë°˜í™˜.
    """
    if videos_df.empty or comments_df.empty:
        print("ì˜ìƒ ë˜ëŠ” ëŒ“ê¸€ ë°ì´í„°ê°€ ë¶€ì¡±í•˜ì—¬ AI ë¶„ì„ì„ ê±´ë„ˆëœë‹ˆë‹¤.")
        return None

    print("\n=================================================================")
    print(" AI ë¶„ì„ ë¦¬í¬íŠ¸ ìƒì„± ì‹œì‘")
    print("=================================================================")
    print("AI ë¶„ì„ìš© ë°ì´í„° ì¤€ë¹„ ì¤‘...")

    # ì˜ìƒ ì •ë³´ë¥¼ í…ìŠ¤íŠ¸ë¡œ ì •ë¦¬
    video_summary = []
    for _, video in videos_df.iterrows():
        summary = f"""ì œëª©: {video['title']}
ì¡°íšŒìˆ˜: {video['view_count']:}íšŒ
ì¢‹ì•„ìš”: {video['like_count']:}ê°œ
ëŒ“ê¸€: {video['comment_count']:}ê°œ
ì—…ë¡œë“œ: {video['upload_date']}
ì¬ìƒì‹œê°„: {video['duration']}"""
        video_summary.append(summary)
    videos_text = '\n\n'.join(video_summary)

    # ëŒ“ê¸€ ì •ë³´ë¥¼ í…ìŠ¤íŠ¸ë¡œ ì •ë¦¬
    comments_for_ai = []

    for _, comment in meaningful_comments.iterrows():
      # ëŒ“ê¸€ì„ ì½ê¸° ì‰½ê²Œ ì •ë¦¬
      comment_line = f"[{comment['like_count']} LIKES ] {comment['author']}: {comment['text']}"
      comments_for_ai.append(comment_line)
    # AIì—ê²Œ ë¶„ì„ì„ ìš”ì²­í•  í”„ë¡¬í”„íŠ¸ ì‘ì„±
    analysis_prompt = f"""YouTube ì±„ë„ '{channel_name}' ë¶„ì„ ë°ì´í„°ì…ë‹ˆë‹¤.
ë°©ì†¡ PDì™€ ì½˜í…ì¸  ê¸°íšìë¥¼ ìœ„í•œ ì‹¤ìš©ì ì¸ ë¶„ì„ì„ í•´ì£¼ì„¸ìš”.

=== ë¶„ì„ ëŒ€ìƒ ===
ì±„ë„: {channel_name}

=== ì¸ê¸° ì˜ìƒ ì„±ê³¼ ===
{videos_text}

=== ì‹œì²­ì ëŒ“ê¸€ ë°˜ì‘ (ëŒ“ê¸€ ì¢‹ì•„ìš” ìˆœ) ===
{comments_for_ai}

{prompt}
"""

    try:
        print(f"AI ë¶„ì„ ìš”ì²­ ì¤‘... (ëª¨ë¸: {model_name})")
        response = genai_client.models.generate_content(
            model=model_name, # ëª¨ë¸ ì´ë¦„ í˜•ì‹ì— ë§ê²Œ ìˆ˜ì •
            contents=analysis_prompt
        )
        ai_analysis = response.text

        print("âœ… AI ë¶„ì„ ì™„ë£Œ!")
        print("\n" + "="*60)
        print("AI ë¶„ì„ ê²°ê³¼")
        print("="*60)
        print(ai_analysis)
        print("="*60)

        # AI ë¶„ì„ ê²°ê³¼ë¥¼ íŒŒì¼ë¡œ ì €ì¥
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        analysis_filename = f"ai_analysis_{channel_name}_{timestamp}.txt"
        analysis_path = os.path.join(save_path, analysis_filename)

        with open(analysis_path, 'w', encoding='utf-8') as f:
            f.write(f"=== {channel_name} AI ë¶„ì„ ë¦¬í¬íŠ¸ ===\n")
            f.write(f"ë¶„ì„ ì¼ì‹œ: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
            f.write(f"ë¶„ì„ ì˜ìƒ: {len(videos_df)}ê°œ\n")
            f.write(f"ë¶„ì„ ëŒ“ê¸€: {len(comments_df)}ê°œ\n")
            f.write(f"AI ì „ë‹¬ ëŒ“ê¸€: {len(meaningful_comments)}ê°œ\n\n")
            f.write("="*60 + "\n")
            f.write(ai_analysis)

        print(f"\nâœ… AI ë¶„ì„ ê²°ê³¼ ì €ì¥ ì™„ë£Œ: {analysis_path}")
        return analysis_path

    except Exception as e:
        print(f"âŒ AI ë¶„ì„ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        print("API í‚¤ë¥¼ í™•ì¸í•˜ê±°ë‚˜ ë‚˜ì¤‘ì— ë‹¤ì‹œ ì‹œë„í•´ì£¼ì„¸ìš”.")
        return None


def retrieve_channel_videos(channel_name,  sort_by='popular', start_date=None, end_date=None, limit=MAX_VIDEOS):
  """
  ìœ íŠœë¸Œ ë°ì´í„° APIë¥¼ í™œìš©í•˜ì—¬ ë¹„ë””ì˜¤ ë°ì´í„°ë¥¼ ìˆ˜ì§‘í•©ë‹ˆë‹¤.
  Args:
      channel_name (str): ë¶„ì„í•  ì±„ë„ì˜ ì´ë¦„
      sort_by (str): 'popular' (ì¸ê¸°ìˆœ), 'latest' (ìµœì‹ ìˆœ) ì¤‘ í•˜ë‚˜
      start_date (str): 'YYYY-MM-DD' í˜•ì‹ì˜ ì‹œì‘ ë‚ ì§œ
      end_date (str): 'YYYY-MM-DD' í˜•ì‹ì˜ ì¢…ë£Œ ë‚ ì§œ
      limit (int): ê°€ì ¸ì˜¬ ë¹„ë””ì˜¤ ê°œìˆ˜

  Returns:
      pd.DataFrame: ìˆ˜ì§‘ëœ ë¹„ë””ì˜¤ ì •ë³´ê°€ ë‹´ê¸´ ë°ì´í„°í”„ë ˆì„
  """
  channel_id = get_channel_id_by_handle(channel_name)
  videos = get_all_channel_videos(youtube, channel_id)
  videos = analyze_channel_videos(videos, sort_by, start_date, end_date, limit)
  videos_df = convert_dataframe_from_video_data(videos)
  return videos_df

def save_raw_data(videos_df, comments_df):
  """
  ìˆ˜ì§‘í•œ ë¹„ë””ì˜¤ ë° ëŒ“ê¸€ ë°ì´í„°ë¥¼ êµ¬ê¸€ ë“œë¼ì´ë¸Œì— ì €ì¥í•©ë‹ˆë‹¤.

   Args:
        videos_df: ë¹„ë””ì˜¤ì˜ ë°ì´í„°í”„ë ˆì„
        comments_df: ëŒ“ê¸€ì˜ ë°ì´í„°í”„ë ˆì„
  """
  timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
  if not videos_df.empty:
    video_filename = f"videos_{timestamp}.csv"
    video_path = os.path.join(SAVE_PATH, video_filename)
    # í•œê¸€ì´ ê¹¨ì§€ì§€ ì•Šë„ë¡ utf-8-sig ì¸ì½”ë”© ì‚¬ìš©
    videos_df.to_csv(video_path, index=False, encoding='utf-8-sig')
    print(f"ì˜ìƒ ë°ì´í„° ì €ì¥: {video_filename}")

  # ëŒ“ê¸€ ì •ë³´ ì €ì¥ (CSV íŒŒì¼ë¡œ)
  if not comments_df.empty:
      comments_filename = f"comments_{timestamp}.csv"
      comments_path = os.path.join(SAVE_PATH, comments_filename)
      comments_df.to_csv(comments_path, index=False, encoding='utf-8-sig')
      print(f"ëŒ“ê¸€ ë°ì´í„° ì €ì¥: {comments_filename}")

  print(f"ì €ì¥ ì™„ë£Œ! íŒŒì¼ ìœ„ì¹˜: {SAVE_PATH}")
```
