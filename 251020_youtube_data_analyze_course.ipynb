```python
# =================================================================
# 1ë‹¨ê³„: ì¸ê¸° ì˜ìƒ ì •ë³´ ìˆ˜ì§‘í•˜ê¸° (yt-dlp ì‚¬ìš©)
# =================================================================

# ìˆ˜ì§‘ëœ ì˜ìƒ ì •ë³´ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
videos_data = []

try:
    # 1ë‹¨ê³„: ì±„ë„ì˜ ì˜ìƒ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    print("ì±„ë„ ì˜ìƒ ëª©ë¡ ê°€ì ¸ì˜¤ëŠ” ì¤‘...")

    # yt-dlp ëª…ë ¹ì–´ ì‹¤í–‰ (ì±„ë„ì˜ ì˜ìƒ ëª©ë¡ì„ JSONìœ¼ë¡œ ê°€ì ¸ì˜¤ê¸°)
    cmd = [
        'yt-dlp',                                    # yt-dlp í”„ë¡œê·¸ë¨ ì‹¤í–‰
        '--flat-playlist',                           # ì˜ìƒ ëª©ë¡ë§Œ ê°€ì ¸ì˜¤ê¸° (ë‹¤ìš´ë¡œë“œ ì•ˆí•¨)
        '--dump-json',                               # ì •ë³´ë¥¼ JSON í˜•íƒœë¡œ ì¶œë ¥
        '--playlist-end', str(MAX_VIDEOS * 2),       # ì—¬ìœ ë¶„ìœ¼ë¡œ ë” ë§ì´ ê°€ì ¸ì˜¤ê¸°
        f'https://www.youtube.com/{CHANNEL_NAME}/videos'  # ì±„ë„ URL
    ]

    # ëª…ë ¹ì–´ ì‹¤í–‰ (ìµœëŒ€ 60ì´ˆ ëŒ€ê¸°)
    result = subprocess.run(cmd, capture_output=True, text=True, timeout=60)

    # ì‹¤í–‰ ê²°ê³¼ í™•ì¸
    if result.returncode != 0:  # ì˜¤ë¥˜ ë°œìƒ
        print(f"ì˜ìƒ ëª©ë¡ ê°€ì ¸ì˜¤ê¸° ì‹¤íŒ¨: {result.stderr}")
    else:
        print("ì˜ìƒ ëª©ë¡ ê°€ì ¸ì˜¤ê¸° ì„±ê³µ!")

        # ê²°ê³¼ì—ì„œ ì˜ìƒ ID ì¶”ì¶œ
        video_ids = []
        for line in result.stdout.strip().split('\n'):
            if line.strip():  # ë¹ˆ ì¤„ì´ ì•„ë‹ˆë©´
                try:
                    video_info = json.loads(line)  # JSONìœ¼ë¡œ ë³€í™˜
                    video_id = video_info.get('id', '')  # ì˜ìƒ ID ì¶”ì¶œ
                    if video_id:
                        video_ids.append(video_id)
                except:
                    continue  # ë³€í™˜ ì‹¤íŒ¨í•˜ë©´ ë„˜ì–´ê°€ê¸°

        print(f"ì°¾ì€ ì˜ìƒ ê°œìˆ˜: {len(video_ids)}ê°œ")

        # 2ë‹¨ê³„: ê° ì˜ìƒì˜ ìƒì„¸ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
        for i, video_id in enumerate(video_ids):
            # ì›í•˜ëŠ” ê°œìˆ˜ë§Œí¼ ìˆ˜ì§‘í–ˆìœ¼ë©´ ì¤‘ë‹¨
            if len(videos_data) >= MAX_VIDEOS:
                break

            print(f"ì˜ìƒ ì •ë³´ ìˆ˜ì§‘ ì¤‘... [{i+1}/{min(len(video_ids), MAX_VIDEOS*2)}]")

            try:
                # ê°œë³„ ì˜ìƒ ìƒì„¸ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
                detail_cmd = [
                    'yt-dlp',
                    '--dump-json',           # JSON ì •ë³´ë§Œ ê°€ì ¸ì˜¤ê¸°
                    '--skip-download',       # ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì•ˆí•¨
                    f'https://www.youtube.com/watch?v={video_id}'  # ì˜ìƒ URL
                ]

                # ëª…ë ¹ì–´ ì‹¤í–‰ (ìµœëŒ€ 30ì´ˆ ëŒ€ê¸°)
                detail_result = subprocess.run(detail_cmd, capture_output=True, text=True, timeout=30)

                if detail_result.returncode == 0:  # ì„±ê³µ
                    video_detail = json.loads(detail_result.stdout)  # JSON ë³€í™˜

                    # ì—…ë¡œë“œ ë‚ ì§œ í™•ì¸ (ë¶„ì„ ê¸°ê°„ì— í•´ë‹¹í•˜ëŠ”ì§€ ì²´í¬)
                    upload_date_str = video_detail.get('upload_date', '')

                    # ì¡°íšŒìˆ˜ í™•ì¸ (ê¸°ì¤€ ë¯¸ë‹¬ì´ë©´ ê±´ë„ˆë›°ê¸°)
                    view_count = video_detail.get('view_count', 0)
                    if view_count < VIEW_THRESHOLD:
                        print(f"  ì¡°íšŒìˆ˜ ë¶€ì¡±: {view_count:,}íšŒ")
                        continue

                    # ì˜ìƒ ì •ë³´ ì •ë¦¬í•´ì„œ ì €ì¥
                    video_data = {
                        'video_id': video_detail.get('id', ''),                    # ì˜ìƒ ID
                        'title': video_detail.get('title', ''),                   # ì œëª©
                        'channel_title': video_detail.get('uploader', ''),        # ì±„ë„ëª…
                        'upload_date': upload_date_str,                           # ì—…ë¡œë“œ ë‚ ì§œ
                        'view_count': view_count,                                 # ì¡°íšŒìˆ˜
                        'like_count': video_detail.get('like_count', 0),          # ì¢‹ì•„ìš” ìˆ˜
                        'comment_count': video_detail.get('comment_count', 0),    # ëŒ“ê¸€ ìˆ˜
                        'duration': video_detail.get('duration', 0),             # ì¬ìƒì‹œê°„(ì´ˆ)
                        'duration_string': video_detail.get('duration_string', ''), # ì¬ìƒì‹œê°„(ë¬¸ì)
                        'description': video_detail.get('description', '')[:200], # ì„¤ëª…(ì•ë¶€ë¶„ë§Œ)
                        'tags': ','.join(video_detail.get('tags', [])),          # íƒœê·¸ë“¤
                        'url': video_detail.get('webpage_url', ''),              # ì˜ìƒ URL
                        'collected_at': datetime.now().isoformat()               # ìˆ˜ì§‘ ì‹œê°„
                    }

                    videos_data.append(video_data)
                    print(f"  ìˆ˜ì§‘ ì™„ë£Œ: '{video_data['title'][:30]}...' - {view_count:,}íšŒ")

                # ì„œë²„ì— ë¶€ë‹´ ì£¼ì§€ ì•Šê¸° ìœ„í•´ 1ì´ˆ ì‰¬ê¸°
                time.sleep(1)

            except subprocess.TimeoutExpired:
                print(f"  ì‹œê°„ ì´ˆê³¼: {video_id}")
                continue
            except Exception as e:
                print(f"  ì˜¤ë¥˜ ë°œìƒ: {e}")
                continue

except subprocess.TimeoutExpired:
    print("ì˜ìƒ ëª©ë¡ ê°€ì ¸ì˜¤ê¸° ì‹œê°„ ì´ˆê³¼")
except Exception as e:
    print(f"ì˜ìƒ ì •ë³´ ìˆ˜ì§‘ ì¤‘ ì˜¤ë¥˜: {e}")

# ìˆ˜ì§‘ ê²°ê³¼ë¥¼ pandas ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜ (ì—‘ì…€ í‘œì™€ ë¹„ìŠ·)
videos_df = pd.DataFrame(videos_data)

print(f"\nìˆ˜ì§‘ ì™„ë£Œ! ì´ {len(videos_df)}ê°œ ì˜ìƒ ì •ë³´ ìˆ˜ì§‘")

# ìˆ˜ì§‘í•œ ì˜ìƒ ëª©ë¡ ì¶œë ¥
if not videos_df.empty:
    print("\nìˆ˜ì§‘ëœ ì˜ìƒ ëª©ë¡:")
    for idx, row in videos_df.iterrows():
        print(f"  {idx+1}. {row['title'][:40]}... - {row['view_count']:,}íšŒ")
else:
    print("ì¡°ê±´ì— ë§ëŠ” ì˜ìƒì´ ì—†ìŠµë‹ˆë‹¤.")
    print(f"(ìµœì†Œ ì¡°íšŒìˆ˜ {VIEW_THRESHOLD:,}íšŒ)")
```

## ìµœì¢… ê²°ê³¼ë¬¼:
ì´ ë‹¨ê³„ê°€ ì™„ë£Œë˜ë©´, ì±„ë„ì˜ ì£¼ìš” ì˜ìƒì— ëŒ€í•œ ì‹œì²­ìë“¤ì˜ êµ¬ì²´ì ì¸ ì˜ê²¬ì´ ë‹´ê¸´ **'ëŒ“ê¸€ ë°ì´í„°'**ê°€ `comments_df`ë¼ëŠ” ì´ë¦„ì˜ ì •í˜•í™”ëœ ë°ì´í„°í”„ë ˆì„(ì—‘ì…€ í‘œ í˜•ì‹)ìœ¼ë¡œ ìƒì„±ë©ë‹ˆë‹¤. ì´ ë°ì´í„°ëŠ” AIë¥¼ í™œìš©í•œ ì‹¬ì¸µ ë¶„ì„ ë‹¨ê³„ì˜ ê°€ì¥ í•µì‹¬ì ì¸ ì¬ë£Œê°€ ë©ë‹ˆë‹¤.

```python
# =================================================================
# 2ë‹¨ê³„: ì˜ìƒ ëŒ“ê¸€ ìˆ˜ì§‘í•˜ê¸° (YouTube Data API ì‚¬ìš©)
# =================================================================

# ìˆ˜ì§‘ëœ ëŒ“ê¸€ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
all_comments = []

# ìˆ˜ì§‘í•œ ì˜ìƒì´ ìˆì„ ë•Œë§Œ ëŒ“ê¸€ ìˆ˜ì§‘
if not videos_df.empty:
    for idx, video in videos_df.iterrows():
        video_id = video['video_id']
        video_title = video['title']

        print(f"\n[{idx+1}/{len(videos_df)}] '{video_title[:40]}...' ëŒ“ê¸€ ìˆ˜ì§‘ ì¤‘...")

        try:
            # YouTube Data APIë¥¼ ì‚¬ìš©í•´ì„œ ëŒ“ê¸€ ê°€ì ¸ì˜¤ê¸°
            comments_request = youtube.commentThreads().list(
                part='snippet',                    # ê¸°ë³¸ ì •ë³´ë§Œ ê°€ì ¸ì˜¤ê¸°
                videoId=video_id,                  # ì˜ìƒ ID
                maxResults=MAX_COMMENTS_PER_VIDEO, # ê°€ì ¸ì˜¬ ëŒ“ê¸€ ìˆ˜
                order='relevance'                  # ì¸ê¸°ìˆœìœ¼ë¡œ ì •ë ¬
            )

            # API ìš”ì²­ ì‹¤í–‰
            comments_response = comments_request.execute()

            # ì´ ì˜ìƒì˜ ëŒ“ê¸€ë“¤ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
            video_comments = []

            # ê° ëŒ“ê¸€ ì •ë³´ ì²˜ë¦¬
            for item in comments_response['items']:
                # ëŒ“ê¸€ì˜ ê¸°ë³¸ ì •ë³´ ì¶”ì¶œ
                top_comment = item['snippet']['topLevelComment']['snippet']

                # ëŒ“ê¸€ ì •ë³´ ì •ë¦¬
                comment_data = {
                    'video_id': video_id,                                      # ì˜ìƒ ID
                    'video_title': video_title,                                # ì˜ìƒ ì œëª©
                    'comment_id': item['snippet']['topLevelComment']['id'],    # ëŒ“ê¸€ ID
                    'text': top_comment['textDisplay'],                        # ëŒ“ê¸€ ë‚´ìš©
                    'author': top_comment['authorDisplayName'],                # ì‘ì„±ì ì´ë¦„
                    'like_count': top_comment.get('likeCount', 0),             # ëŒ“ê¸€ ì¢‹ì•„ìš” ìˆ˜
                    'published_at': top_comment['publishedAt'],                # ì‘ì„± ì‹œê°„
                    'reply_count': item['snippet'].get('totalReplyCount', 0),  # ë‹µê¸€ ìˆ˜
                    'collected_at': datetime.now().isoformat()                 # ìˆ˜ì§‘ ì‹œê°„
                }

                video_comments.append(comment_data)

            # ì´ ì˜ìƒì˜ ëŒ“ê¸€ë“¤ì„ ì „ì²´ ëŒ“ê¸€ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            all_comments.extend(video_comments)
            print(f"  {len(video_comments)}ê°œ ëŒ“ê¸€ ìˆ˜ì§‘ ì™„ë£Œ")

            # API ì‚¬ìš©ëŸ‰ ë³´í˜¸ë¥¼ ìœ„í•´ 0.5ì´ˆ ì‰¬ê¸°
            time.sleep(0.5)

        except HttpError as e:
            error_message = str(e)

            if 'commentsDisabled' in error_message:
                print("  ëŒ“ê¸€ì´ ë¹„í™œì„±í™”ëœ ì˜ìƒì…ë‹ˆë‹¤")
            elif 'quotaExceeded' in error_message:
                print("  API ì‚¬ìš©ëŸ‰ ì´ˆê³¼! ëŒ“ê¸€ ìˆ˜ì§‘ì„ ì¤‘ë‹¨í•©ë‹ˆë‹¤")
                break
            else:
                print(f"  API ì˜¤ë¥˜: {error_message}")

        except Exception as e:
            print(f"  ëŒ“ê¸€ ìˆ˜ì§‘ ì˜¤ë¥˜: {e}")

else:
    print("ë¶„ì„í•  ì˜ìƒì´ ì—†ì–´ ëŒ“ê¸€ ìˆ˜ì§‘ì„ ê±´ë„ˆëœë‹ˆë‹¤")

# ëŒ“ê¸€ ë°ì´í„°ë¥¼ pandas ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
comments_df = pd.DataFrame(all_comments)

print(f"\nëŒ“ê¸€ ìˆ˜ì§‘ ì™„ë£Œ! ì´ {len(comments_df)}ê°œ ëŒ“ê¸€ ìˆ˜ì§‘")

# ëŒ“ê¸€ ìˆ˜ì§‘ ê²°ê³¼ ì¶œë ¥
if not comments_df.empty:
    print(f"ê³ ìœ  ì‘ì„±ì: {comments_df['author'].nunique()}ëª…")
    print(f"í‰ê·  ì¢‹ì•„ìš”: {comments_df['like_count'].mean():.1f}ê°œ")

    print("\nì¸ê¸° ëŒ“ê¸€ ìƒ˜í”Œ:")
    # ì¢‹ì•„ìš”ê°€ ë§ì€ ëŒ“ê¸€ 3ê°œ ì¶œë ¥
    top_comments = comments_df.nlargest(3, 'like_count')
    for idx, row in top_comments.iterrows():
        # ëŒ“ê¸€ì´ ê¸¸ë©´ ì•ë¶€ë¶„ë§Œ ì¶œë ¥
        text_preview = row['text'][:50] + '...' if len(row['text']) > 50 else row['text']
        print(f"  {row['author']}: {text_preview} (ì¢‹ì•„ìš” {row['like_count']}ê°œ)")
```
